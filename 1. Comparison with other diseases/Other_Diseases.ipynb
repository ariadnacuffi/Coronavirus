Lately I have seen people try to minimize COV-19 by comparing it to the Swine Flu. 
https://www.propublica.org/article/no-president-trump-the-coronavirus-is-nothing-like-h1n1-swine-flu-either
Althoug it is true that so far, Swine Flu has killed 30x more people, we need more context here. 

```python
import pandas as pd

fig = go.Figure()
fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Cholera'],
                name="Cholera",
                line_color='deepskyblue',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['COVID-19'],
                name="COVID-19",
                line_color='orange',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Ebola'],
                name="Ebola",
                line_color='darkblue',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Swine Flu'],
                name="Swine Flu",
                line_color='hotpink',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['SARS'],
                name="SARS",
                line_color='olive',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Measles 2019 - now'],
                name="Measles 2019 - now",
                line_color='purple',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Measles 2011-14'],
                name="Measles 2011-14",
                line_color='red',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Cholera 2008-09'],
                name="Cholera 2008-09",
                line_color='yellow',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Cholera 2016-now'],
                name="Cholera 2016-now",
                line_color='yellowgreen',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Ebola.1'],
                name="Ebola.1",
                line_color='magenta',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Swine Flu.1'],
                name="Swine Flu.1",
                line_color='lavender',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['Meningitis'],
                name="Meningitis",
                line_color='chocolate',
                opacity=0.8))

fig.add_trace(go.Scatter(
                x=deaths_by_day['DAY'],
                y=deaths_by_day['MERS'],
                name="MERS",
                line_color='lightpink',
                opacity=0.8))

fig.show()
```
